```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.coef_*3300+reg.intercept_
#y=mx+b
```

```python
#plot the scatter plot with the slope

%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
d =pd.read_csv("areas.csv")
d
```

```python
p= reg.predict(d)
```

```python
d['prices'] = p  #create a column in teh areas.csv the values are from model preeidction
```

```python
d.to_csv("prediction.csv", index=False)
```
